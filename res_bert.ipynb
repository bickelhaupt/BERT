```python
def acc(df):
    df = df.rename(columns = {"Unnamed: 0" : "index", "pred_sentiment" : "pred", "delta":"real"})
    df = df[['real', 'pred']]
    df.real = df.real*100
    df.pred = df.pred
    df.plot()
     
acc(d_AAL)    
acc(d_GME)
acc(d_SPY)
```
